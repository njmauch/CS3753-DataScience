```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn import datasets, linear_model
from scipy.stats import binom

years = np.array([2005, 2006, 2007, 2008, 2009])
sales = np.array([12, 19, 29, 37, 45])
plt.scatter(years, sales)
model = linear_model.LinearRegression()
model.fit(years.reshape(-1,1),sales.reshape(-1,1))
print(model.coef_)
print(model.intercept_)
```

```python
def gradient_descent(x,y):
    beta = alpha = 0    
    num_iterations = 10000
    rate = 0.01
    n = len(x)
    for i in range(num_iterations):
        mse = (1/n)*((y - (beta + (alpha * x))).dot(y- (beta + (alpha * x))))
        #d_slope = (-2/n)*(x.dot(y - (beta + (alpha * x))))
        d_slope = ((years - years.mean()).dot(sales - sales.mean()))
        d_slope = -(d_slope/((years - years.mean()).dot((years - years.mean()))))
        #d_intercept = (-2/n)*np.sum(y - (beta + (alpha * x)))
        d_intercept = sales.mean() - (d_slope * years.mean())
        alpha = alpha - (rate * d_slope)
        beta = beta - (rate * d_intercept)
        print("iteration= {}, slope= {}, intercept= {},mse= {}".format(i, alpha, beta, mse))       
```

```python
gradient_descent(years, sales)
```

```python
import numpy.random as rn
import numpy as np
_tosses = 16
_size = 10**5
nHeads = rn.uniform(0.0, 1.0,size=(_size, _tosses))>0.5
nHeads = nHeads.sum(1)
plt.hist(nHeads, bins=range(18))
plt.xlabel("nHeads")
plt.ylabel("Number of coins")
plt.title("Q2a: Histogram")
plt.show()
```

```python
counts = plt.hist(nHeads, bins=range(18), density=True)
bin_center = (counts[1][1:]+counts[1][:-1])/2
plt.xlabel("k")
plt.ylabel("P(nHeads = k)")
plt.title("Q2b: Probability Mass Function")
plt.bar(bin_center, counts[0]/sum(counts[0]))
plt.show()
```

```python
sorted_nHeads = np.sort(nHeads)
counts, bin_edges = np.histogram(sorted_nHeads, bins=range(18))
prob_heads = np.cumsum(counts)/len(nHeads)
plt.plot(bin_edges[1:], prob_heads)
plt.xlabel("k")
plt.ylabel("P(nHeads <= k)")
plt.title("Q2c: Cumulative Distribution Function")
plt.show()
```

```python
prob_binom_heads = binom.cdf(range(17), 16, p=0.5)
plt.xlabel("Emperical CDF")
plt.ylabel("Theoretical CDF")
plt.title("Q2d: scatter plot")
plt.scatter(prob_heads, prob_binom_heads)
plt.show()
```

```python
plt.xlabel("k")
plt.ylabel("CDF")
plt.title("Q2d: line plot")
plt.plot(range(17), prob_heads, color='blue', marker='o', linestyle='solid')
plt.plot(range(17), prob_binom_heads, color='orange', marker='o', linestyle='dotted')
plt.legend(['Emperical CDF', 'Theoretical CDF'])
plt.show()
```

```python
plt.xlabel("Emperical CDF")
plt.ylabel("Theoretical CDF")
plt.title("Q2d: loglog plot")
plt.loglog(prob_heads, prob_binom_heads, ".")
plt.show()
```
